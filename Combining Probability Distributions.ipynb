```python
import numpy as np
from matplotlib.ticker import FuncFormatter
```

# Creating a Uniform Probability Distribution

First we want to create a uniform probability distribution, in this example lets create a distribution with a minimum value of 100, a maximum value of 200, is equally likely to be any value in between the min and max, and cannot be a value outside of that range. Our mean value will thus be 150.

```python
#Create a array of zeros and ones
start = np.zeros(100)
middle = start + 1
end = start
step_function = np.concatenate((start, middle, end))

#Create a function for plotting
def plot_my_probability(array):
    plt.plot(array)
    plt.ylim(-.1*np.max(array),1.3*np.max(array))

#Plot my data
plot_my_probability(step_function)
plt.ylabel("Likelihood")
plt.xlabel("Value")
plt.show()
```

# Now Lets Add them together!

When you want to add two probability distributions you do something called "[convolving](https://en.wikipedia.org/wiki/Convolution_of_probability_distributions)". Let's see what happens when we convolve uniform probability function with itself over and over!

```python
iterations = 6

convolved = step_function

for i in range(0,iterations):
    #Do the convolution!!!
    convolved = np.convolve(convolved, step_function)
    
    #Make the Plots look Good
    plt.subplot(iterations, 1, i+1)
    plot_my_probability(convolved)
    plt.xlim(0,len(convolved))
    plt.title("{} Distributions".format(i+2))
    plt.ylabel("Likelihood")
    plt.xlabel("Value")
    
#Some more plotting stuff
plt.tight_layout(rect=[0, 0, 0.6, iterations/2])
plt.show()
```

# Conclusions

We can see that as we add more uniform distributions together, the net probability distribution quickly approachs a normal probability distribution. (This is called the "[central limit theorm](https://en.wikipedia.org/wiki/Central_limit_theorem)")

Once we have added 7 distributions together, the mean value of the final distribution is 7x150 = 1,050 which intuitivley makes sense (150 was the original mean value of our uniform distribution)

We can also see that once we have added 7 distributions together the likelihood of reaching the absolute minimum possible value (7x100 = 700), and maximum possible value (7x200 = 1400) is quite small
